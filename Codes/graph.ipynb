```python
import matplotlib.pyplot as plt
import networkx as nx
import pandas as pd
```

```python
path = r'C:\Users\rishe\Desktop\DMproject\topicGraph.csv'
```

```python
df = pd.read_csv(path).values
df
```

```python
edge = []
for row in range(len(df)):
    if row%2 == 0:
        l = []
        l.append(df[row][0])
        l.append(df[row][1])
        edge.append(l)
edge
```

```python
G = nx.Graph()
G.add_edges_from(edge)
pos = nx.spring_layout(G)
f = plt.figure()
f.set_figwidth(4000)
f.set_figheight(6000)
```

```python
G.nodes()
```

```python
nx.draw(
    G, pos, edge_color='yellow', width=1, linewidths=1,
    node_size=1000, node_color='pink', alpha=0.9,
    labels={node: node for node in G.nodes()}
)
```
