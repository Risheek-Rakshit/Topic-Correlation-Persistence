```python
def filesArray(s):
    t = s[1:-1]
    t = t.replace("'","")
    t = t.split(", ")

    return t
```

```python
def topicsArray(s):
    t = s[1:-1]
    t = t.split("], ")
    # print("t" , t)
    k = []
    for i in t:
        i = i.replace("[","")
        i = i.replace(']',"")
        i = i.split(", ")

        k.append(i)
    return k
```

```python
pathCorrelations = r'C:\Users\rishe\Desktop\DMproject2\Correlation-deepscore\\'
```

```python
def expand(file):
    file = str(file)
    if len(file) == 7:
        return str(file)
    else:
        num = 7-len(file)
        for i in range(0,num):
            file = str(0) + file
        return str(file)
```

```python
def fetchCorrelation(start,topic1,destination,topic2):
    try:
        path = pathCorrelations + expand(start) + '-' + expand(destination) + '.csv'
        value = pd.read_csv(path).values
        row = int(topic1)
        column = int(topic2)+1
        return value[row][column]
    except:
        path =  pathCorrelations + expand(destination) + '-' + expand(start) + '.csv'
        value = pd.read_csv(path).values
        row = int(topic2)
        column = int(topic1)+1
        return value[row][column]
```

```python
def writeNetwork(file1,topic1,file2,topic2):
    lout = []

    score = fetchCorrelation(file1,topic1,file2,topic2)
    entry1 = str(file1) + '-' + str(topic1)
    entry2 = str(file2) + '-' + str(topic2)
    lout.append(entry1)
    lout.append(entry2)
    lout.append(score)

  #print(lout)
  #cleaned=pd.DataFrame(lout)
    fd = r'C:\Users\rishe\Desktop\DMproject2\TopicGraphs\topicGraph.csv'
    with open(fd,'a') as f:
        write = csv.writer(f)
        write.writerow(lout)
```

```python
k = []
missed = 0
for row in range(len(paths)):
    if row%2==1:
        continue
  #print(paths[row][0])
    files = filesArray(paths[row][0])
    topics = topicsArray(paths[row][1])
    left = 0
    right = 1
    while right<len(files):
        for j in range(len(topics)):
            if int(topics[j][left]) == -1 or int(topics[j][right])== -1:
                missed = missed+1
                continue
            writeNetwork(files[left],topics[j][left],files[right],topics[j][right])
            #k.append(str(files[left] + "-" + topics[j][left] + "|" + files[right] + "-" + topics[j][right]))
        left = left+1
        right = right+1
```
