```python
import re
import nltk
import sklearn
import os
import numpy as np
import pandas as pd
```

```python
from numpy import argmax
from sklearn.metrics.pairwise import cosine_similarity
from collections import defaultdict
from scipy import spatial
from sentence_transformers import SentenceTransformer
sbert_model = SentenceTransformer('bert-base-nli-mean-tokens')
```

```python
datapath = r'C:\Users\rishe\Desktop\DMproject\ExtractedTopics'
dirList = os.listdir(datapath)
dirList
```

```python
len(dirList)
```

```python
count = 0
for file in dirList:
    path = datapath + '\\' + file
    dataset = pd.read_csv(path).values
    textContent = []
    for i in range(len(dataset)):
        textContent.append(dataset[i][2])
    sentence_embeddings = sbert_model.encode(textContent,convert_to_numpy=True)

    cleaned=pd.DataFrame(sentence_embeddings)
    fd = r'C:\Users\rishe\Desktop\DMproject\Bert\\'
    fd = fd + file
    with open(fd,'w') as f:
        cleaned.to_csv(f)
        
    count = count+1
    if count%1000 == 0:
        print(count)
```
