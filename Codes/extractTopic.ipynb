```python
import pandas as pd
import re
import os
```

```python
path = r'C:\Users\rishe\Desktop\DMproject\Topics'
dir1 = os.listdir(path)
len(dir1)
```

```python
path = str(r'C:\Users\rishe\Desktop\DMproject\Topics\\')
count = 0
for file in dir1:
    npath = path + file
    dataset=pd.read_csv(npath).values
    topics=[]
    for iter in range(len(dataset)):
        curr=[]
        mystring=dataset[iter][2][3:]
        update=re.sub('[^A-Za-z]+', ' ', mystring)
        curr.append(dataset[iter][1])
        curr.append(update)
        topics.append(curr)

    cleaned=pd.DataFrame(topics)
    fd = r'C:\Users\rishe\Desktop\DMproject\ExtractedTopics\\'
    fd = fd + file
    with open(fd,'w') as f:
        cleaned.to_csv(f)
        
    count = count+1
    if count%1000 == 0:
        print(count)
```
